```python
#Extracting the dataset
import pandas as pd
data = pd.read_csv("/content/spam.csv", encoding='ISO-8859-1')
data
```

```python
#Removing columns with NaN values
data.dropna()
```

```python
#Removing unecessary columns from the dataset
data.drop(columns=['Unnamed: 3', 'Unnamed: 2', 'Unnamed: 4'], axis = 1, inplace = True)
data
```

```python
#To indicate the no.of non-spam messages which is 86%
print(data.v1.value_counts(normalize=True))
```

```python
#Splitting the data into training and testing sets
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
X_train, X_test, y_train, y_test = train_test_split(data['v2'], data['v1'], test_size=0.2, random_state=42)
```

```python
# TF-IDF Vectorization - converting the text data into numerical format
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

```python
# Model training and evaluation - Naive Bayes
from sklearn.naive_bayes import MultinomialNB
nb_classifier = MultinomialNB()
nb_classifier.fit(X_train_tfidf, y_train)
nb_predictions = nb_classifier.predict(X_test_tfidf)
```

```python
from sklearn.metrics import accuracy_score
print("Naive Bayes: ")
print("Accuracy: ", accuracy_score(y_test, nb_predictions))
```

Through Naive Bayes the accuracy comes out to be 96%

```python
# Model training and evaluation - Support Vector Machines
from sklearn.svm import SVC
classifier=SVC(kernel='linear',random_state=0)
classifier.fit(X_train_tfidf,y_train)
```

```python
y1_predict=classifier.predict(X_test_tfidf)
y1_predict
```

```python
from sklearn import metrics
print("Accuracy: ")
print(accuracy_score(y_test,y1_predict))
```

Therefore we can see that using Support Vector Machine, the prediction comes out to be with a 97% accuracy



```python
from sklearn.linear_model import LogisticRegression
lr_classifier = LogisticRegression()
lr_classifier.fit(X_train_tfidf, y_train)
lr_predictions = lr_classifier.predict(X_test_tfidf)
```

```python
print("Logistic Regression:")
print("Accuracy:", accuracy_score(y_test, lr_predictions))
```

Through LR, the accuracy comes out to be 95%
